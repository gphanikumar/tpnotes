## Flow fields in cylindrical coordinate system

We use cyclindrical coordinate system to represent some of the planar flows easier.

```python
E.<x,y> = EuclideanSpace()
```

```python
cartesian = E.cartesian_coordinates()
```

```python
polar.<r,th> = E.polar_coordinates()
```

```python
polar_frame = E.polar_frame()
cartesian_frame = E.cartesian_frame()
```

```python
E.coord_change(polar, cartesian).display()
```

```python
E.coord_change(cartesian, polar).display()
```

```python
E.set_default_frame(polar_frame)
```

```python
P = E.scalar_field({polar: -th}, name='P')
#P = E.scalar_field({polar: cos(th)/r}, name='P')
P.display()
```

```python
from sage.manifolds.operators import grad
```

```python
V = -grad(P)
```

```python
V.display(polar_frame,polar)
```

```python
w3f = E.vector_field(V[1], V[2], name='w3f')
```

```python
w3f.display(polar_frame, polar)
```

```python
w3f.display(cartesian_frame, cartesian)
```

```python
Parallelism().set(nproc=4)
w3f.plot(chart=cartesian,ranges={x: (0.5,5), y: (0.5,5)}, scale=0.5, arrowsize=3)
```

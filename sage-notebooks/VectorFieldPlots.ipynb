# Vector Field Plots

You can provide functional forms for the u_x and u_y components of the velocity field and visualize the velocity field. Below is an example of a field defined as:

$$ \vec{u} = \left( y^2-25 \right) \, \hat{e}_x + 0 \, \hat{e}_y $$

```python
var('x,y')
plot_vector_field((y^2-25,0),(x,-5,5),(y,-5,5))
```

The magnitude of the unidirectional velocity can be visualized as a parabola.

```python
plot(y^2-25,(y,-5,5))
```

# Velocity field from a potential

One can create a velocity field using gradient of a velocity potential. It is defined as:

$$ \vec{u} = - \vec{\nabla} \Phi $$

```python
f=-1/(x^2+y^2)
f.gradient()
```

```python
plot_vector_field(f.gradient(),(x,1,5),(y,1,5))
```

# Velocity field in polar coordinates

One can use the Manifolds package of sagemath to define velocity fields in other coordinate systems and visualize. one can use this package also to convert the expressions from one coordinate system to another automatically.

Here we visualize a velocity field defined as follows:

$$ \vec{u} = \left( -y \right) \, \hat{e}_x + \left( x \right) \, \hat{e}_y $$

```python
E.<x,y> = EuclideanSpace()
```

```python
v = E.vector_field(-y, x, name='v')
```

```python
v.display()
```

```python
v.plot()
```

```python
cartesian = E.cartesian_coordinates()
polar.<r,ph> = E.polar_coordinates()
```

```python
E.coord_change(polar, cartesian).display()
```

```python
E.coord_change(cartesian, polar).display()
```

```python
polar_frame = E.polar_frame()
cartesian_frame = E.cartesian_frame()
```

Now that the polar and cartesian coordinate frames are defined, we can visualize the velocity field defined as 

$$ \vec{u} = {1 \over r} \, \hat{e}_r $$

```python
E.set_default_frame(polar_frame)
w = E.vector_field(1/r, 0, name='w')
```

```python
w.display(polar_frame, polar)
```

```python
E.set_default_frame(cartesian_frame)
w.display(cartesian_frame)
```

The following cell will take sometime to complete execution.

```python
w.plot(ranges={x: (1,5), y: (1,5)})
```
